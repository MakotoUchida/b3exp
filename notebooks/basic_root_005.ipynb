### MCAのデータからROOTでヒストグラムを作成する方法
MCAによって得られたデータからROOTをもちいてヒストグラムを作成する。  
テキストファイルを読み込んでヒストグラムを生成するimportHist()という関数が準備されている。  
```c++
    TH1D* importHist(string);
```
よって次のように引数にデータファイル名をしていするとヒストグラムが生成される。  
正しくはTH1D* hとすべきだがautoを用いることが自動的にデータ型を判別してくれる。

````c++
auto h = importHist("test.dat");
````

キャンパス作成

````c++
c = new TCanvas();
````

````c++
h->Draw();
c->Draw();
````

もしもイベント数が少なくて１チャンネルあたりのエントリーが少ない時にはbinの幅を変える関数を
使うと良い。
```c++
TH1D* rebin(TH1D* h, int nbin, string name);
```
nbinごとに新たなbinが作られる。
以下の例ではbinの数を1/8にするようにして新しいヒストグラムを作成している。

````c++
auto hRebin = rebin(h,8,"rebinned histogram");
hRebin->Draw();
c->Draw();
````
